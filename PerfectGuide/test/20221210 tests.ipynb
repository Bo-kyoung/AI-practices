```python
data = pd.read_csv('train.csv')
```

```python
#2. 타이타닉 데이터의 ‘age0’ 에 0상수값을 할당하여 column을 추가하고, 이를 없앤다,
data['age0'] = 0
data.head()
```

```python
data.drop('age0',axis = 1, inplace = True)
data.head()
```

```python
#0,1,2 row를 없애본다
data.drop([1,2,3],axis = 0, inplace=True)
data.head()
```

```python
#1번 df의 column을 [’one’,’two’]로 변경한후, (1,col2)를 호출하자 이 데이터 프레임을 list로 만들자
test2 = np.arange(6)
test2 = test2.reshape(2,-1)
print(test2)
test3 = pd.DataFrame(test2, columns = ['one','two','three'])
test3.loc[1,'two']
```

```python
#titanic데이터에서 age가 60이상일때, 이름과 age 출력 (loc를 이용한 방법도 연습)
data[data['Age']>60][['Name']].head(3)
#data[data['Age']>60]['Name'] 이거랑 다름..!
```

```python
#pclass로 groupby해서 age의 mean과 max, count구하기
data.groupby(by = 'Pclass')['Age'].agg(['mean','max'])
```

```python
#a=[1,2,3]에 일괄적으로 x**2 적용하기 (lamda)
a = [1,2,3]
list(map(lambda x: x**2,a)) #list 붙여줘야함
```
